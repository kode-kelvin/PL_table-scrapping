We scrapping of HTML table

Website: https://www.skysports.com/premier-league-table/2019


```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
```

```python
#FIRST WE SCRAPE 2019/2020 TABLE
```

```python

url = "https://www.skysports.com/premier-league-table/2019"
```

```python
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
l = soup.find('table', class_="standing-table__table")
```

```python
print(l)
```

```python
league_tb = []
for team in l.find_all('tbody'):
    rows = team.find_all('tr')
    for row in rows:
        pl_team = row.find('td', class_='standing-table__cell standing-table__cell--name').text.strip()
        pl_points = row.find_all('td', class_='standing-table__cell')[9].text
        pl = row.find_all('td', class_='standing-table__cell')[2].text
        w = row.find_all('td', class_='standing-table__cell')[3].text
        d = row.find_all('td', class_='standing-table__cell')[4].text
        l = row.find_all('td', class_='standing-table__cell')[5].text
        f = row.find_all('td', class_='standing-table__cell')[6].text
        a = row.find_all('td', class_='standing-table__cell')[7].text
        gd = row.find_all('td', class_='standing-table__cell')[8].text
        #print(pl_team, pl_points)
        
        lister = { #creating a dict with data scrapped
            'Clubs':pl_team,
            'Pl':pl,
            'W':w,
            'D':d,
            'L':l,
            'F':f,
            'A':a,
            'GD':gd,
            'Pts': pl_points
            
            
            
            
        }
        
        league_tb.append(lister) #adding it to list
         
table = pd.DataFrame(league_tb)
```
